```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
Xs = np.array([ 0.387, 0.723, 1.00, 1.52, 5.20, 9.54, 19.2, 30.1, 39.5 ])

Ys = np.array([ 0.241, 0.615, 1.00, 1.88, 11.9, 29.5, 84.0, 165.0, 248 ])

N = 9
```

```python
# To two significant figures, what is the correlation coefficient
r = np.round(np.corrcoef(Xs, Ys)[0, 1], 2)
print(r)
```

```python
# draw a scatter plot of the data
plt.scatter(Xs, Ys)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Residuals
residuals = Ys - np.polyval(np.polyfit(Xs, Ys, 1), Xs)
print(residuals)
# draw a scatter plot of the residuals
plt.scatter(Xs, residuals)
plt.show()
```

```python
import statsmodels.api as sm
sm.qqplot(Xs, line='s')
plt.title("X distribution")
plt.show()
```

```python
# Create the Q-Q plots for both the X and the Y values.
sm.qqplot(Ys, line='s')
plt.title("Y distribution")
plt.show()
```
